Se comienza importando la cantidad de librerías necesarias para hacer EDA y ETL a los datos.

```python
import pandas as pd
import numpy as np
```

Realizamos la ingesta de los archivos csv y json.

```python
amazondf = pd.read_csv (r'amazon_prime_titles.csv')
# Se comienza el proceso de EDA visualizando los datos de cada dataframe.
amazondf.head()
```

```python
#Buscamos y sumamos los valores nulos.
amazondf.isna().sum()
```

```python
#Vemos la cantidad de columnas y filas de cada uno.
amazondf.shape
```

```python
#Identificamos si hay datos duplicados, en caso de ser necesario se eliminan.
amazondf[amazondf.duplicated(keep=False)]
```

```python
#Ningún df contiene filas duplicadas en este caso. Ahora, para la columna duration la separo para
#  dejar por un lado los minutos que dura, y por otro una columna donde diga la cantidad 
# de temporadas "seasons", y debajo se visualizan los dataframes para ver que todo vaya bien.
amazondf['min']= amazondf['duration'].apply(lambda i: i.split()[0] 
    if i.split()[1]=='min' 
    else '0')

amazondf['seasons']= amazondf['duration'].apply(lambda i: i.split()[0] 
    if i.split(" ",1)[1]=='Seasons' or i.split(" ",1)[1]=='Season' 
    else '0')

amazondf.head(4)
```

```python
#Lo siguiente es agregar una columna a cada dataframe con su respectiva plataforma, 
# para luego concatenarlos y que cada Movie o Tv Show tenga a qué plataforma pertenece.
amazondf['platform'] = 'Amazon'
amazondf.head(4)
```

```python
#Ya que la columna duration quedó separada, se elimina porque no va a servir más.
amazondf.drop(['duration'], axis=1)
```

```python
#Ahora se debe cambiar el tipo de dato de las columnas 'tiempo_dur' y 'seasons', 
# a integer para que se puedan realizar las consultas.
amazondf['seasons'] = amazondf['seasons'].astype(int)
amazondf['min'] = amazondf['min'].astype(int)
```

```python
amazondf.dtypes
```

```python
#Por último, a los valores nulos se les completa con el string 'Sin dato',
#  para que quede completo el dataframe.
```

```python
amazondf.fillna('Sin dato', inplace=True)
amazondf.sample(4)
```

```python
#Al dataframe final, lo convertimos a un csv para trabajarlo en nuestro archivo main.py .
amazondf = pd.DataFrame(amazondf)
amazondf.to_csv('df.csv')
```
